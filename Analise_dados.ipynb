```python
from matplotlib import pyplot as plt
import utils
```

```python
def plot_cluster(dados, cluster, func):
    x = []
    y = []
    i = cluster[0][1]

    for d in cluster:
        if(i != d[1]):
            func.scatter(x,y)
            x=[]
            y=[]
            i=d[1]

        for j in dados:
            if(j[0] == d[0]):
                cord = [j[1],j[2]]
        x.append(cord[0])
        y.append(cord[1])

    func.scatter(x,y)
```

```python
dados =  utils.readTableTxt(open("./datasets/c2ds1-2sp.txt"))
cluster = []

cluster.append(utils.readTableTxt(open("./datasets/c2ds1-2spReal.clu")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/particao_kmedias.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/single_link_k_2.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/single_link_k_3.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/single_link_k_4.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/single_link_k_5.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/complete_link_k_2.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/complete_link_k_3.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/complete_link_k_4.txt")))
cluster.append(utils.readTableTxt(open("./c2ds1-2sp/complete_link_k_5.txt")))
                            
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(min(5*2, 12), 3*1), constrained_layout=True)
fig.suptitle("c2ds1-2sp", fontsize=16)
ax[0].set_title("Partição Real")
ax[1].set_title("Partição obtida com k-médias")


plot_cluster(dados, cluster[0], ax[0])
plot_cluster(dados, cluster[1], ax[1])

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(min(5*4, 12), 3*2), constrained_layout=True)
ax[0][0].set_title("Obtida com single-link k=2")
ax[0][1].set_title("Obtida com single-link k=3")
ax[0][2].set_title("Obtida com single-link k=4")
ax[0][3].set_title("Obtida com single-link k=5")


plot_cluster(dados, cluster[2], ax[0][0])
plot_cluster(dados, cluster[3], ax[0][1])
plot_cluster(dados, cluster[4], ax[0][2])
plot_cluster(dados, cluster[5], ax[0][3])

ax[1][0].set_title("Obtida com complete-link k=2")
ax[1][1].set_title("Obtida com complete-link k=3")
ax[1][2].set_title("Obtida com complete-link k=4")
ax[1][3].set_title("Obtida com complete-link k=5")


plot_cluster(dados, cluster[6], ax[1][0])
plot_cluster(dados, cluster[7], ax[1][1])
plot_cluster(dados, cluster[8], ax[1][2])
plot_cluster(dados, cluster[9], ax[1][3])

```
