```python
from matplotlib import pyplot as plt
import utils
```

```python
def plot_particao(dados, particao, func):
    qtd_clusters = 0
    for obj in particao:
        if(obj[1] > qtd_clusters):
            qtd_clusters = int(obj[1])
    x = []
    y = []
    for i in range(0, qtd_clusters + 1):
        x.append([])
        y.append([])

    for obj in particao:
        for j in dados:
            if(j[0] == obj[0]):
                cord = [j[1],j[2]]
        x[int(obj[1])].append(cord[0])
        y[int(obj[1])].append(cord[1])
    
    for i in range(0, qtd_clusters + 1):
        func.scatter(x[i],y[i])
```
